create a consolidate dataframe for further analyses after filtering to keep only information that's relevant

```python
print(dataframe_NewYork.shape)
print(dataframe_Seattle.shape)
print(dataframe_Shanghai.shape)
print(dataframe_SF.shape)

df_analysis=dataframe_NewYork.append([dataframe_Shanghai, dataframe_SF, dataframe_Seattle])
```

```python
# keep only columns that include venue name, and anything that is associated with location
filtered_columns = ['name', 'categories'] + [col for col in df_analysis.columns if col.startswith('location.')] + ['id']+['area']
df_filtered = df_analysis.loc[:, filtered_columns]

# function that extracts the category of the venue
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']

# filter the category for each row
df_filtered['categories'] = df_filtered.apply(get_category_type, axis=1)

# clean column names by keeping only last term
df_filtered.columns = [column.split('.')[-1] for column in df_filtered.columns]

df_filtered
```

identify variety of cushine in each city

```python
summary=df_filtered.groupby(['categories','area']).size().reset_index(name='counts')
```

```python
summary.sort_values(by=['area','counts'], ascending=False)
```
